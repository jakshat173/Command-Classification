```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns;sns.set()
import nltk
from nltk import word_tokenize
from nltk.corpus import stopwords
import string
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv("/Users/akshatjain/Desktop/dataset.csv")
print(data.head())
```

```python
data.shape
```

```python
data['Category'].value_counts()
```

```python
def text_cleaning(a):
    remove_punctuation=[char for char in a if char not in string.punctuation]
    #print(remove_punctuation)
    remove_punctuation=''.join(remove_punctuation)
    #print(remove_punctuation)
    return [word for word in remove_punctuation.split() if word.lower() not in stopwords.words('english')]
```

```python
print(data['Command'].apply(text_cleaning))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
bow=CountVectorizer(analyzer=text_cleaning).fit(data['Command'])

bow.vocabulary_
```

```python
command_bow=bow.transform(data['Command'])
print(command_bow)
```

```python
x=command_bow.toarray()
print(x)
x.shape
```

```python
from sklearn.feature_extraction.text import TfidfTransformer

tfidf=TfidfTransformer().fit(command_bow)
print(tfidf)

command_tfidf=tfidf.transform(command_bow)
print(command_tfidf)
```

```python
from sklearn.model_selection import train_test_split

y=data['Category']
X=command_tfidf

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

```

```python
from sklearn.naive_bayes import MultinomialNB

model=MultinomialNB().fit(X_train,y_train )
y_predict=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix


print(accuracy_score(y_test, y_predict))
confusion_matrix(y_test, y_predict)
```
